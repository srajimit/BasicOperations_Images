<h4> <font color = "red"> 1. Load Necessary Libraries </font></h4>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

<h4> <font color = "red"> 2. Read the image in Grey Scale Mode </font></h4>

```python
#Read the image
image = cv2.imread("bricks.jpeg",0)
```

<h4> <font color = "red"> 3. Compute the histogram </font></h4>

```python
# Compute the histogram
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
```

<h4> <font color = "red"> 4. Equalize the image </font></h4>

```python
# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(image)
```

```python
# Display the image and histogram
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Sample Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Histogram ')
plt.plot(hist)
plt.xlim([0, 256])

plt.subplot(1, 3, 3)
plt.title("Equalized Image")
plt.imshow(equalized_image, cmap='gray')



plt.show()
```
