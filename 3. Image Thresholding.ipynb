```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img6 = cv2.imread("crossword.jpg",0)
show_pic(img6)
```

```python
th2 = cv2.adaptiveThreshold(img6,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
th3 = cv2.adaptiveThreshold(img6,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,15,8)
```

```python
show_pic(th3)
```

```python
blended = cv2.addWeighted(src1=th2,alpha=0.7,src2=th3,beta=0.3,gamma=0)
show_pic(blended)
```
