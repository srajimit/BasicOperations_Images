```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Define the input matrices
img1 = np.array([
    [10, 20, 30],
    [15, 25, 35],
    [20, 30, 40]
], dtype=np.uint8)  # Ensure the type is uint8 for OpenCV compatibility

img2 = np.array([
    [70, 60, 50],
    [65, 55, 45],
    [60, 50, 40]
], dtype=np.uint8)  # Ensure the type is uint8 for OpenCV compatibility

# Set the alpha and beta values for blending
alpha = 0.6
beta = 0.2
# Perform the weighted blending using OpenCV's addWeighted function
blended_img = cv2.addWeighted(img1, alpha, img2, beta, 0)

# Display the images
plt.figure(figsize=(9, 3))

# Display img1
plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray', vmin=0, vmax=100)
plt.title('Image 1')
plt.axis('off')

# Display img2
plt.subplot(1, 3, 2)
plt.imshow(img2, cmap='gray', vmin=0, vmax=100)
plt.title('Image 2')
plt.axis('off')

# Display blended image
plt.subplot(1, 3, 3)
plt.imshow(blended_img, cmap='gray', vmin=0, vmax=100)
plt.title('Blended Image')
plt.axis('off')

# Show all plots
plt.tight_layout()
plt.show()

print("Blended Image:\n", blended_img)
```

<p><font color = "tomato"> Creating Region of Interest </font></p>
<ul>
    <li> get the shape of large image </li>
    <li> get the shape of small image </li>
    <li> find the position where to place the image </li>
    <li> <font color = blue> top left corner: roi = img1[0:rows, 0:cols ] # TOP LEFT CORNER 
        where rows represent height of small image, cols represent width of small image. img1 is the large image </font> </li>
    <li> <font color = "purple">bottom right corner: img1[y_offset:H,x_offset:W] 
        where H, W represents height and width of large image "img 1"
        start at: x_offset = width of large image- width of small image; y_offset =  height of large image - height of small image </font> </li>
        

```python
#get the shape of the resized img 5 - bigger image
img5_rs.shape
```

```python
#get the shape of the resized img 6 - smaller image
img6_rs.shape
```

```python
#get the postion where to place smaller image on bigger image
x_offset = 1200 - 600 # width 600 -- column
y_offset = 1200 - 610 #height 590 --row
```

```python
#To create ROI of the same shape
rows,cols,chans = img6_rs.shape
roi = img5_rs[y_offset:1200,x_offset:1200]
#roi = img5_rs[0:rows,0:cols]
plt.imshow(roi)
```

```python
#shape of ROI
roi.shape
```

<p> <font color = "orange"> Steps for creating mask</font>
    <ul>
        <li> convert the image to grey </li>
        <li> create thresholding or white background using full() </li>
        <li> create inverse mask </li>
        </ul>
</p>

<h3><font color="blue"> Convert Image to Grayscale (cv2.cvtColor) </font> <h3>

```python
#step 1 Converting image to gray
img6gray = cv2.cvtColor(img6_rs,cv2.COLOR_BGR2GRAY)
```

```python
img6gray
```

```python
img6gray.shape
```

```python
plt.imshow(img6gray,cmap='gray')
```

<h3><font color="blue"> Invert the grey scale image </font> <h3>

```python
mask_inv = cv2.bitwise_not(img6gray)
```

```python
mask_inv
```

```python
plt.imshow(mask_inv,cmap='gray')
```

```python
mask_inv.shape
```

<h3><font color="blue"> Create a White Background (cv2.cvtColor) </font> <h3>

```python
white_background = np.full(img6_rs.shape, 255, dtype=np.uint8)
white_background.shape
```

```python
white_background
```

```python
mask_inv
```

<h3><font color="blue"> Apply Mask to White Background </font> <h3>

```python
bk = cv2.bitwise_or(white_background, white_background, mask=mask_inv)
bk
```

```python
plt.imshow(bk)
```

```python
plt.imshow(mask_inv,cmap='gray')
```

<h3><font color="blue"> Extract the foreground </font> <h3>

```python
foreg = cv2.bitwise_or(img6_rs,img6_rs,mask = mask_inv)
```

```python
plt.imshow(foreg,cmap='gray')
```

```python
foreg.shape
```

<h3><font color="blue"> Apply roi(extracted from large image) and foreground (extracted in small image)</font> <h3>

```python
Final_roi = cv2.bitwise_or(roi,foreg)
```

```python
plt.imshow(Final_roi)
```

<h3><font color="blue"> Place the final roi on the portion of larger image</font> <h3>


```python
#add the image in the rest of the image
large_img = img5_rs
small_img = Final_roi

large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]] = small_img
#large_img[y_offset:y_offset+small_img.shape[0],x_offset:x_offset+small_img.shape[1]] = small_img
```
