<h2> <font color="tomato"> Load necessary Libraries </font></h2>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h2> <font color="tomato"> Load the image </font></h2>

<p> <font color = "purple">
    <ul>
        <li> Reads an image from the file bricks.jpeg.</li>
        <li> Converts the pixel values to floating-point numbers. </li>
        <li> Normalizes the pixel values to a range between 0 and 1. </li>
    </ul>
    </p>

```python
def load_img():
    img = cv2.imread('bricks.jpeg').astype(np.float32) / 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

<h2> <font color="tomato"> Display the image </font></h2>

<p> <font color = "purple">
    <ul>
<li> plt.figure(...): This function creates a new figure (a blank canvas) on which plots can be drawn. It initializes the figure for plotting. </li>
<li> figsize=(12,10): The figsize parameter specifies the size of the figure in inches. In this case, the figure will be 12 inches wide and 10 inches tall.</li> </ul> </font>
    </p>

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img1 = load_img()
display_img(img1)
```

<h2> <font color="tomato"> Gamma Correction for brightening or darkening the image </font></h2>

```python
img1 = load_img()
gamma = 0.3 # gamma can be from 0.1 ; no 0 or negative value
#gamma_img1 = np.array(255 * (img1 / 255) ** gamma, dtype='uint8')
gamma_image1 = np.power(img1, gamma)
display_img(gamma_image1)
```

```python
img2 = load_img()
gamma = 3
gamma_img2 = np.power(img2,gamma)
display_img(gamma_img2)
```

<h2> <font color="tomato"> Blurring </font></h2>

<h2> <font color="purple"> 1. Using Kernel </font></h2>

```python
kernel_= np.ones(shape=(5,5),dtype=np.float32)/12 # create any kernel, here it is divided by 12 (it can be changed)
```

```python
kernel_
```

```python
kernel_img = cv2.filter2D(img1,ddepth=-1,kernel = kernel_)
```

```python
display_img(kernel_img)
```

```python
box_mean_ = cv2.blur(img1,ksize=(3,3))
```

```python
display_img(box_mean_)
```

```python
gaussian_blurred_img= cv2.GaussianBlur(img1,(5,5),10)
display_img(gaussian_blurred_img)
```

```python
median_blur_ = cv2.medianBlur(img1,5)
display_img(median_blur_)
```
